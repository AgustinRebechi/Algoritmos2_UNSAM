## Ejercicio 2

 Implemente una clase Monedero que permita gestionar la cantidad de dinero que
una persona dispone en un momento dado. La clase deberá tener un constructor
que permitirá crear un monedero con una cantidad de dinero inicial y deberá
definir un método para meter dinero en el monedero, otro para sacarlo y
finalmente, otro para consultar el disponible; solo podrá conocerse la cantidad de
dinero del monedero a través de este último método. Por supuesto, no se podrá
sacar más dinero del que haya en un momento dado en el monedero

```python
class Monedero:

    def __init__(self, cantidad=0):
        self.cantidad = cantidad

    def ingresar(self, valor):
        self.cantidad += valor
        print(f'se ingreso {valor} pesimios. Saldo actual {self.cantidad} pesimios')

    def retirar(self, valor):

        if valor <= self.cantidad:
            self.cantidad -= valor
            print(f'se retiro {valor} pesimios. Saldo actual {self.cantidad} pesimios')
        else:
            print(f'No se puede retirar ese monto de pesimios, saldo insuficiente')

    def mostrar(self):
        print(f'monto actual: {self.cantidad} pesimios')

billetera = Monedero(100)

billetera.ingresar(100)
billetera.mostrar()

billetera.retirar(50.5)
billetera.mostrar()

billetera.retirar(200)
billetera.mostrar()

billetera.retirar(149.50)
billetera.mostrar()


```
